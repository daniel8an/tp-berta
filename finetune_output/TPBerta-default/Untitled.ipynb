```python
def plot_arr(arr_):
    # Plotting the test_metric array
    plt.figure(figsize=(10, 6))
    plt.plot(arr_, marker='o', linestyle='-', color='b', label='Test Metric')

    # Adding labels and title
    plt.xlabel('Epoch')
    plt.ylabel('Test Metric')
    plt.title('Test Metric Over Epochs')
    plt.legend()

    # Display the plot
    plt.show()
```

```python
plot_arr(data['test_metric'])
```

```python
plot_arr(data_default['test_metric'])
```
